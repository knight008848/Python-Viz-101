```python
import matplotlib.pyplot as plt
import numpy as np

# 为了让图表在Notebook中直接显示（如果你用Jupyter），需要这行魔法命令
# %matplotlib inline
print("✅ 库导入成功！舞台准备就绪。")

# 1. 定义数据
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']  # 横坐标：星期
latte_sales = [65, 70, 80, 90, 85, 95, 75]  # “醇香拿铁”每日销量
americano_sales = [50, 55, 60, 70, 65, 80, 60]  # “经典美式”每日销量

# 计算每日总销量（为折线图准备）
total_sales = np.array(latte_sales) + np.array(americano_sales)

print("📊 数据准备完成！")
print("日期：", days)
print("拿铁销量：", latte_sales)
print("美式销量：", americano_sales)
print("每日总销量：", total_sales)

# 【第2步代码】创建一张画布(fig)和一个坐标系(ax)
# figsize 控制图表大小（宽，高，单位英寸）
fig, ax = plt.subplots(figsize=(10, 6))
print(f"🎨 画布创建成功！坐标系对象是：{ax}")

# 【第3步代码】设置柱状图的宽度并绘制
bar_width = 0.35
# 生成每组柱子的中心x轴位置
x = np.arange(len(days))

# 绘制“醇香拿铁”的柱子，颜色设为拿铁色，标签为'Latte'
bars1 = ax.bar(x - bar_width/2, latte_sales, bar_width,
               label='Latte', color='sandybrown', edgecolor='black')
# 绘制“经典美式”的柱子，颜色设为咖啡色，位置在拿铁柱子右边
bars2 = ax.bar(x + bar_width/2, americano_sales, bar_width,
               label='Americano', color='sienna', edgecolor='black')
print("📊 柱状图绘制完成！")

# 【第4步代码】绘制总销量折线图，用红色圆点虚线表示，线宽加粗
# 注意：折线图的x轴位置直接用x（即柱状图中心），而不是x +/- bar_width/2
line = ax.plot(x, total_sales, color='firebrick', marker='o', linestyle='--',
               linewidth=2, label='Total Sales', markersize=8)
print("📈 折线图叠加完成！")

# 【第4步代码】设置图表修饰元素
# 1. 设置主标题和坐标轴标签
ax.set_title('Weekly Coffee Sales Report', fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Day of the Week', fontsize=12)
ax.set_ylabel('Number of Cups Sold', fontsize=12)

# 2. 设置x轴的刻度标签（用我们定义的days替换掉数字0,1,2...）
ax.set_xticks(x) # 设置刻度位置
ax.set_xticklabels(days) # 设置刻度标签

# 3. 显示图例（自动识别我们在bar和plot时设置的label）
ax.legend()

# 4. 显示网格线（让读数更清晰），设置网格线为浅灰色虚线
ax.grid(True, linestyle='--', alpha=0.6)

# 5. (可选) 为每个柱子顶部添加数值标签，更直观
def autolabel(bars):
    """在柱子顶部显示数值"""
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)
autolabel(bars1)
autolabel(bars2)

print("✨ 所有图表修饰元素添加完成！")

#【第6步代码】
# 自动调整布局，防止标签被裁剪
plt.tight_layout()

# 显示图表
plt.show()

# 将图表保存为高清PNG图片（可用于报告或PPT）
fig.savefig('weekly_coffee_sales.png', dpi=300, bbox_inches='tight')
print("💾 图表已保存为 'weekly_coffee_sales.png'！")
```
