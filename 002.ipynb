```python
import seaborn as sns
import matplotlib.pyplot as plt

# 一行代码开启全局美颜，后续所有图表自动变好看！
sns.set_theme(style="whitegrid")

# Seaborn 内置 17 个经典数据集，直接用！
tips = sns.load_dataset("tips")
print(tips.head())

# 画直方图 + 密度曲线，就这一行！
sns.histplot(data=tips, x="total_bill", kde=True, color="mediumpurple")

# 简单加个标题
plt.title("Distribution of Total Bill", fontsize=14)
plt.xlabel("Total Bill ($)")
plt.ylabel("Count")

plt.show()
```

```python
# 画箱线图，按性别分组，还是只要一行！
sns.boxplot(data=tips, x="sex", y="tip", hue="sex", palette="Set2")

plt.title("Tip Distribution by Gender", fontsize=14)

plt.show()
```
