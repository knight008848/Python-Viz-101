```python
# 📦 1. 环境准备与数据加载
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 通过 rc 参数设置中文字体（Seaborn 推荐做法）
sns.set_theme(
    style="whitegrid",
    font_scale=1.1,
    rc={
        "font.sans-serif": ["Microsoft YaHei"],
        "font.family": "sans-serif",
        "axes.unicode_minus": False
    }
)

# 加载并清洗企鹅数据
df = sns.load_dataset("penguins").dropna(subset=["body_mass_g"])

# 快速查看物种分布
print(df["species"].value_counts())
```

```python
# 📊 2. 基础：一行代码画出分组柱状图
sns.barplot(data=df, x="species", y="body_mass_g",
            hue="species", palette="Set2", legend=False)

plt.title("不同种类企鹅平均体重", fontsize=14, pad=15)
plt.xlabel("物种", fontsize=12)
plt.ylabel("体重 (g)", fontsize=12)

plt.show()
```

```python
# 🏷️ 3. 进阶：双剑合璧——Matplotlib 柱顶打标签
fig, ax = plt.subplots(figsize=(8, 6))

# Seaborn 画主体，绑定到 ax
ax = sns.barplot(data=df, x="species", y="body_mass_g",
                 hue="species", palette="Set2", ax=ax,
                 legend=False, errorbar="sd")

ax.set_title("不同种类企鹅平均体重", fontsize=14, pad=15)
ax.set_xlabel("物种", fontsize=12)
ax.set_ylabel("体重 (g)", fontsize=12)

# 核心武器：遍历所有 containers 打标签
for container in ax.containers:
    ax.bar_label(container, fmt="%.0f g", fontsize=11, padding=5)

ax.yaxis.grid(True, linestyle="--", alpha=0.6)
ax.xaxis.grid(False)

plt.tight_layout()
plt.show()
```

```python
# 🚀 4. 高级玩法：多维度对比（按性别分组）
fig, ax = plt.subplots(figsize=(10, 6))

# hue="sex" 双重分组
ax = sns.barplot(data=df, x="species", y="body_mass_g",
                 hue="sex", palette="pastel", ax=ax, errorbar="sd")

ax.set_title("不同种类与性别企鹅平均体重对比", fontsize=14, pad=15)
ax.set_xlabel("物种", fontsize=12)
ax.set_ylabel("体重 (g)", fontsize=12)

# 给两组柱子分别打标签
for container in ax.containers:
    ax.bar_label(container, fmt="%.0f g", fontsize=10, padding=3)

# 图例外置，避免遮挡图表
ax.legend(title="性别", bbox_to_anchor=(1.05, 1), loc="upper left")

ax.yaxis.grid(True, linestyle="--", alpha=0.5)
ax.xaxis.grid(False)

plt.tight_layout()
plt.show()
```

```python
# 💾 5. 完整代码模板（含高清导出）

fig, ax = plt.subplots(figsize=(10, 6))

ax = sns.barplot(data=df, x="species", y="body_mass_g",
                 hue="sex", palette="Set2", ax=ax, errorbar="sd")

ax.set_title("不同种类与性别企鹅平均体重对比", fontsize=14, pad=15)
ax.set_xlabel("物种", fontsize=12)
ax.set_ylabel("体重 (g)", fontsize=12)

for container in ax.containers:
    ax.bar_label(container, fmt="%.0f g", fontsize=10, padding=5)

ax.legend(title="性别", bbox_to_anchor=(1.05, 1), loc="upper left")
ax.yaxis.grid(True, linestyle="--", alpha=0.5)
ax.xaxis.grid(False)

plt.tight_layout()

# 导出高清 PNG (DPI 300)
fig.savefig("penguin_weight_comparison.png", dpi=300, bbox_inches="tight")
print("✅ 图表已保存为 'penguin_weight_comparison.png'")

plt.show()
```
