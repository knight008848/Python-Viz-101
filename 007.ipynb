```python
# 📦 1. 环境准备与数据加载
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 通过 rc 参数设置中文字体(Seaborn 推荐做法)
sns.set_theme(
    style="whitegrid",
    font_scale=1.1,
    rc={
        "font.sans-serif": ["Microsoft YaHei"],
        "font.family": "sans-serif",
        "axes.unicode_minus": False
    }
)

# 加载 mpg 数据集:汽车燃油经济性数据
df = sns.load_dataset("mpg").dropna()

# 快速查看数据结构
print(df.head())
print(f"\n数据规模:{df.shape}")
```

```python
# 🔵 2. 基础散点图:马力 vs 油耗
fig, ax = plt.subplots(figsize=(10, 6))

# Seaborn 一行代码画出散点图
sns.scatterplot(
    data=df,
    x="horsepower",
    y="mpg",
    color="steelblue",
    alpha=0.7,
    s=80,
    ax=ax
)

# Matplotlib 精细美化
ax.set_title("汽车马力与油耗关系", fontsize=16, fontweight="bold", pad=15)
ax.set_xlabel("马力 (HP)", fontsize=12)
ax.set_ylabel("每加仑英里数 (MPG)", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# 🎈 3. 气泡图:加入重量作为第三维度
fig, ax = plt.subplots(figsize=(10, 6))

# size 参数映射重量,sizes 控制气泡大小范围
sns.scatterplot(
    data=df,
    x="horsepower",
    y="mpg",
    size="weight",
    sizes=(50, 400),
    alpha=0.7,
    ax=ax,
    legend=False
)

# Matplotlib 精细美化
ax.set_title("汽车马力、油耗与重量关系(气泡图)", fontsize=16, fontweight="bold", pad=15)
ax.set_xlabel("马力 (HP)", fontsize=12)
ax.set_ylabel("每加仑英里数 (MPG)", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# 🎨 4. 多维散点图:颜色映射产地 + 大形状映射气缸数
fig, ax = plt.subplots(figsize=(12, 7))

sns.scatterplot(
    data=df,
    x="horsepower",
    y="mpg",
    hue="origin",
    style="origin",
    size="weight",
    sizes=(50, 400),
    alpha=0.7,
    ax=ax
)

# Matplotlib 精细美化
ax.set_title("汽车性能多维对比:马力、油耗、重量、产地", fontsize=15, fontweight="bold", pad=15)
ax.set_xlabel("马力 (HP)", fontsize=12)
ax.set_ylabel("每加仑英里数 (MPG)", fontsize=12)

# 图例优化:分开放置,避免遮挡
ax.legend(title="产地", bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# 🎨 5. 自定义高颜值图例
from matplotlib.lines import Line2D

fig, ax = plt.subplots(figsize=(12, 7))

# 画散点图(关闭默认图例)
sns.scatterplot(
    data=df,
    x="horsepower",
    y="mpg",
    hue="origin",
    style="origin",
    size="weight",
    sizes=(50, 400),
    alpha=0.7,
    palette={"usa": "#e74c3c", "europe": "#3498db", "japan": "#2ecc71"},
    ax=ax,
    legend=False
)

# 手动创建高颜值图例
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='美系车 (USA)',
           markerfacecolor='#e74c3c', markersize=12),
    Line2D([0], [0], marker='^', color='w', label='欧系车 (Europe)',
           markerfacecolor='#3498db', markersize=12),
    Line2D([0], [0], marker='s', color='w', label='日系车 (Japan)',
           markerfacecolor='#2ecc71', markersize=12)
]

ax.legend(
    handles=legend_elements,
    title="产地",
    loc="upper right",
    fontsize=11,
    title_fontsize=12,
    frameon=True,
    facecolor="white",
    framealpha=0.9,
    edgecolor="gray"
)

ax.set_title("汽车性能多维对比(自定义图例)", fontsize=15, fontweight="bold", pad=15)
ax.set_xlabel("马力 (HP)", fontsize=12)
ax.set_ylabel("每加仑英里数 (MPG)", fontsize=12)

plt.tight_layout()
plt.savefig("mpg_scatter_advanced.png", dpi=300, bbox_inches="tight")
print("✅ 图表已保存为 'mpg_scatter_advanced.png'")
plt.show()
```

```python
# 💾 6. 完整代码模板(含高清导出)
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 设置主题 + 中文支持
sns.set_theme(
    style="whitegrid",
    font_scale=1.1,
    rc={
        "font.sans-serif": ["Microsoft YaHei"],
        "font.family": "sans-serif",
        "axes.unicode_minus": False
    }
)

# 2. 加载数据(可替换为你自己的 CSV)
# df = pd.read_csv("your_data.csv")
df = sns.load_dataset("mpg").dropna()

# 3. 创建画布
fig, ax = plt.subplots(figsize=(10, 6))

# 4. Seaborn 画散点图
sns.scatterplot(
    data=df,
    x="horsepower",
    y="mpg",
    hue="origin",
    style="origin",
    size="weight",
    sizes=(50, 400),
    alpha=0.7,
    ax=ax
)

# 5. Matplotlib 精细美化
ax.set_title("变量关联分析", fontsize=14, fontweight="bold", pad=15)
ax.set_xlabel("X 轴变量", fontsize=12)
ax.set_ylabel("Y 轴变量", fontsize=12)

# 图例优化
ax.legend(title="分类", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()

# 6. 导出与显示
fig.savefig("scatter_plot.png", dpi=300, bbox_inches="tight")
print("✅ 图表已保存为 'scatter_plot.png'")
plt.show()
```
