## 🪄 二、三步出图：Seaborn 极速实战

### 📌 第 1 步：导入库 + 加载数据

这次我们使用机器学习界最经典的数据集 —— **`iris`**（鸢尾花）！

```python
# 📦 1. 环境准备与数据加载
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 通过 rc 参数设置中文字体
sns.set_theme(
    style="whitegrid",
    font_scale=1.1,
    rc={
        "font.sans-serif": ["Microsoft YaHei"],
        "font.family": "sans-serif",
        "axes.unicode_minus": False
    }
)

# 加载 iris 数据集：三种鸢尾花的形态测量
df = sns.load_dataset("iris")

# 快速查看数据结构
print(df.head())
print(f"\n数据规模：{df.shape}")
print(f"\n物种分类：{df['species'].unique()}")
print(f"\n每个物种的样本数：")
print(df["species"].value_counts())
```

> **💡 数据集简介：**
> `iris` 是统计学和机器学习中最经典的数据集，包含 **150 朵鸢尾花（3 个物种 × 50 朵）** 的形态测量数据：
> - `sepal_length`：花萼长度（cm）
> - `sepal_width`：花萼宽度（cm）
> - `petal_length`：花瓣长度（cm）
> - `petal_width`：花瓣宽度（cm）
> - `species`：物种名称（setosa / versicolor / virginica）
>
> 这是理解分类算法和组间对比的最佳入门数据集！

### 📌 第 2 步：一行代码画出蜂群图

```python
# 🐝 2. 基础蜂群图：每个数据点都不重叠
fig, ax = plt.subplots(figsize=(10, 6))

# Seaborn 一行代码画出蜂群图
sns.swarmplot(
    data=df,
    x="species",        # x 轴：物种分类
    y="petal_length",   # y 轴：花瓣长度
    hue="species",      # 颜色：按物种分组
    palette="Set2",
    size=6,             # 点的大小（调小避免重叠）
    edgecolor="auto",   # 自动匹配边框颜色
    linewidth=0.5,      # 边框线宽
    ax=ax
)

# Matplotlib 精细美化
ax.set_title("三种鸢尾花的花瓣长度对比：蜂群图", fontsize=16, fontweight="bold", pad=15)
ax.set_xlabel("物种", fontsize=12)
ax.set_ylabel("花瓣长度 (cm)", fontsize=12)

plt.tight_layout()
plt.savefig("swarm_basic.png", dpi=300, bbox_inches="tight")
print("✅ 蜂群图已保存为 'swarm_basic.png'")
plt.show()
```

**✅ 完成！**

你没有看错，核心代码就这几行：
* `sns.swarmplot()` 自动将每个数据点排列，**避免重叠**
* `hue="species"` 用颜色区分三个物种
* `size=8` 控制点的大小，`edgecolor` 让边界更清晰
* 点的**密集程度**和**分布形状**一目了然

> **🤔 蜂群图 vs 普通散点图的区别？**
> 
> 普通散点图（stripplot）：点会重叠，小数据集看不出样本量，大数据集成一团黑。
>
> **蜂群图（swarmplot）**：算法自动沿 x 轴方向偏移每个点，保证**不重叠**。
> 就像蜜蜂筑巢一样，每个个体都有位置，整体形成有机轮廓。
>
> **⚠️ 注意：** 蜂群图适合 **≤ 500 个样本**的场景。数据量太大时计算会变慢，此时应改用 `stripplot` 抽样显示。

> 💡 **金句时刻：**
> *"均值告诉你'中心在哪'，蜂群图告诉你'真相是什么'。"*

### 📌 第 3 步：双剑合璧——蜂群图 + 箱线图叠加

```python
# 🐝 + 📦 3. 蜂群图 + 箱线图叠加
fig, ax = plt.subplots(figsize=(10, 6))

# 先画蜂群图（底层）
sns.swarmplot(
    data=df,
    x="species",
    y="petal_length",
    hue="species",
    palette="Set2",
    size=5,             # 调小避免重叠警告
    edgecolor="auto",   # 自动匹配边框颜色
    linewidth=0.5,
    alpha=0.7,
    ax=ax
)

# 再叠加箱线图（顶层）
sns.boxplot(
    data=df,
    x="species",
    y="petal_length",
    color="white",
    width=0.5,
    showcaps=False,
    boxprops={
        "facecolor": "none",
        "edgecolor": "black",
        "linewidth": 2
    },
    whiskerprops={"linewidth": 2, "color": "black"},
    medianprops={"linewidth": 2.5, "color": "red"},
    ax=ax
)

# Matplotlib 精细美化
ax.set_title("鸢尾花花瓣长度：蜂群图 + 箱线图叠加", fontsize=16, fontweight="bold", pad=15)
ax.set_xlabel("物种", fontsize=12)
ax.set_ylabel("花瓣长度 (cm)", fontsize=12)

# 隐藏重复的图例（安全写法）
legend = ax.get_legend()
if legend is not None:
    legend.remove()

plt.tight_layout()
plt.savefig("swarm_boxplot_overlay.png", dpi=300, bbox_inches="tight")
print("✅ 叠加图已保存为 'swarm_boxplot_overlay.png'")
plt.show()
```

**✅ 搞定！** 现在一张图包含两层信息：

* **底层（半透明彩色点）** = 每个样本的真实位置，看出分布密度
* **顶层（白色箱子 + 红线）** = 中位数、四分位数、须线范围
* **`alpha=0.7`** = 蜂群点半透明，箱子清晰可见但不完全遮挡底层
* **`facecolor="none"`** = 箱子透明填充，只保留边框

> **💡 图层叠加顺序（关键！）：**
> 1. 先画 `swarmplot`（底层散点）
> 2. 再画 `boxplot`（顶层统计）
> 
> 如果顺序反了，箱子会被点覆盖看不见。
> 这就是 Matplotlib 的**图层哲学**——先底层，再顶层！

## 🎨 三、进阶玩法：多维度对比 + 论文级美化

### 📌 场景 1：四维度全面对比——花瓣长度 × 宽度 × 物种

```python
# 🔬 4. 多变量对比：花瓣长度 vs 宽度的蜂群矩阵
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 左图：花瓣长度
sns.swarmplot(
    data=df,
    x="species",
    y="petal_length",
    hue="species",
    palette="Set2",
    size=4,
    edgecolor="auto",
    linewidth=0.5,
    alpha=0.7,
    ax=axes[0]
)

# 叠加箱线图
sns.boxplot(
    data=df,
    x="species",
    y="petal_length",
    color="white",
    width=0.5,
    showcaps=False,
    boxprops={"facecolor": "none", "edgecolor": "black", "linewidth": 2},
    whiskerprops={"linewidth": 2, "color": "black"},
    medianprops={"linewidth": 2.5, "color": "red"},
    ax=axes[0]
)

axes[0].set_title("花瓣长度对比", fontsize=14, fontweight="bold")
axes[0].set_xlabel("物种", fontsize=11)
axes[0].set_ylabel("花瓣长度 (cm)", fontsize=11)

# 右图：花瓣宽度
sns.swarmplot(
    data=df,
    x="species",
    y="petal_width",
    hue="species",
    palette="Set2",
    size=4,
    edgecolor="auto",
    linewidth=0.5,
    alpha=0.7,
    ax=axes[1]
)

sns.boxplot(
    data=df,
    x="species",
    y="petal_width",
    color="white",
    width=0.5,
    showcaps=False,
    boxprops={"facecolor": "none", "edgecolor": "black", "linewidth": 2},
    whiskerprops={"linewidth": 2, "color": "black"},
    medianprops={"linewidth": 2.5, "color": "red"},
    ax=axes[1]
)

axes[1].set_title("花瓣宽度对比", fontsize=14, fontweight="bold")
axes[1].set_xlabel("物种", fontsize=11)
axes[1].set_ylabel("花瓣宽度 (cm)", fontsize=11)

# 全局标题
plt.suptitle("三种鸢尾花花瓣形态全面对比（蜂群图 + 箱线图）",
             fontsize=16, fontweight="bold", y=1.02)

# 隐藏多余图例（安全写法）
for ax in axes:
    legend = ax.get_legend()
    if legend is not None:
        legend.remove()

plt.tight_layout()
plt.savefig("iris_multi_feature_comparison.png", dpi=300, bbox_inches="tight")
print("✅ 多特征对比图已保存")
plt.show()

# 打印统计信息
print("\n📊 各物种花瓣长度统计：")
print(df.groupby("species")["petal_length"].describe()[["mean", "std", "min", "50%", "max"]])
```

**✅ 效果拉满！** 左右两图清晰揭示：
* **setosa（山鸢尾）** = 花瓣最短最窄，分布集中（品种特征稳定）
* **versicolor（杂色鸢尾）** = 中等长度，有一定离散度
* **virginica（维吉尼亚鸢尾）** = 花瓣最长最宽，分布范围最大

> **🤫 科研洞察：**
> 观察蜂群图可以发现：
> - setosa 的花瓣长度/宽度与其他两种**完全不重叠** → 这是最容易被线性分类器识别的特征
> - versicolor 和 virginica 有**部分重叠区域** → 这也是为什么早期分类算法在这两种上容易出错
>
> **图表比任何统计检验都更直观地揭示了分类难度！**

### 📌 场景 2：电商用户评分蜂群图——业务场景实战

```python
# 🛒 5. 电商场景：各产品品类的用户评分分布

# 模拟电商产品评分数据（实际业务中从数据库读取）
np.random.seed(42)

categories = ["电子产品", "服装鞋帽", "食品生鲜", "家居日用", "图书文具"]
n_products = 200

products_df = pd.DataFrame({
    "product_id": range(n_products),
    "category": np.random.choice(categories, n_products),
    "rating": np.round(np.clip(
        np.random.normal(loc=4.0, scale=1.2, size=n_products),
        1, 5  # 评分范围 1-5
    ), 1),
    "review_count": np.random.randint(10, 500, n_products)
})

print("电商产品评分数据预览：")
print(products_df.head())
print(f"\n数据规模：{products_df.shape}")
print(f"\n各品类产品数：")
print(products_df["category"].value_counts())
```

```python
# 🛒 画出电商用户评分蜂群图
fig, ax = plt.subplots(figsize=(12, 6))

# 蜂群图（底层）
sns.swarmplot(
    data=products_df,
    x="category",
    y="rating",
    hue="category",
    palette="pastel",
    size=5,
    edgecolor="auto",
    linewidth=0.5,
    alpha=0.6,
    ax=ax
)

# 箱线图叠加（顶层）
sns.boxplot(
    data=products_df,
    x="category",
    y="rating",
    color="white",
    width=0.4,
    showcaps=False,
    boxprops={"facecolor": "none", "edgecolor": "black", "linewidth": 2},
    whiskerprops={"linewidth": 2, "color": "black"},
    medianprops={"linewidth": 2.5, "color": "red"},
    ax=ax
)

# 添加评分基准线
ax.axhline(y=4.0, color="green", linestyle="--", alpha=0.5, linewidth=1.5, label="优秀线 (4.0)")
ax.axhline(y=3.0, color="orange", linestyle="--", alpha=0.5, linewidth=1.5, label="合格线 (3.0)")

ax.set_title("各产品品类用户评分分布（蜂群图 + 箱线图）",
             fontsize=16, fontweight="bold", pad=15)
ax.set_xlabel("产品品类", fontsize=12)
ax.set_ylabel("用户评分", fontsize=12)
ax.set_ylim(0.5, 5.5)

# 只显示基准线图例（安全写法）
legend = ax.get_legend()
if legend is not None:
    handles, labels = ax.get_legend_handles_labels()
    if len(handles) >= 2:
        ax.legend(handles[-2:], labels[-2:], fontsize=10, loc="lower left")

plt.tight_layout()
plt.savefig("ecommerce_rating_swarm.png", dpi=300, bbox_inches="tight")
print("✅ 电商评分蜂群图已保存")
plt.show()
```

**✅ 效果：** 业务洞察一目了然：
* **每个品类的评价分布** = 蜂群点的疏密显示评分集中度
* **中位数红线** = 快速比较哪个品类评分最高
* **4.0 优秀线 / 3.0 合格线** = 一眼看出哪些品类低于预期
* **离群点（箱子外的点）** = 极端好评或差评的异常产品

> **🤫 业务洞察：**
> 如果某品类的蜂群点**密集在 4.0 以上**，但**中位数低于 4.0**：
> → 说明有少量极差评价拉低了均值，需要调查这些差评产品。
>
> **箱线图只看统计，蜂群图看到真相。**

## 📦 五、完整代码模板（复制即用）

```python
# 💾 完整代码模板：蜂群图 + 箱线图叠加
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 设置主题 + 中文支持
sns.set_theme(
    style="whitegrid",
    font_scale=1.1,
    rc={
        "font.sans-serif": ["Microsoft YaHei"],
        "font.family": "sans-serif",
        "axes.unicode_minus": False
    }
)

# 加载数据（可替换为你自己的 CSV）
# df = pd.read_csv("your_data.csv")
df = sns.load_dataset("iris")

# 创建画布
fig, ax = plt.subplots(figsize=(10, 6))

# 第 1 层：蜂群图（底层散点）
sns.swarmplot(
    data=df,
    x="species",          # x 轴：分类
    y="petal_length",     # y 轴：数值
    hue="species",        # 颜色分组
    palette="Set2",
    size=6,
    edgecolor="auto",
    linewidth=0.5,
    alpha=0.7,            # 透明度，避免遮挡
    ax=ax
)

# 第 2 层：箱线图（顶层统计）
sns.boxplot(
    data=df,
    x="species",
    y="petal_length",
    color="white",
    width=0.5,
    showcaps=False,
    boxprops={"facecolor": "none", "edgecolor": "black", "linewidth": 2},
    whiskerprops={"linewidth": 2, "color": "black"},
    medianprops={"linewidth": 2.5, "color": "red"},
    ax=ax
)

# Matplotlib 精细美化
ax.set_title("分类变量对比：蜂群图 + 箱线图", fontsize=16, fontweight="bold", pad=15)
ax.set_xlabel("分类变量", fontsize=12)
ax.set_ylabel("数值", fontsize=12)

# 清理多余图例（安全写法）
legend = ax.get_legend()
if legend is not None:
    legend.remove()

# 导出高清图
plt.tight_layout()
fig.savefig("swarm_boxplot.png", dpi=300, bbox_inches="tight")
print("✅ 图表已保存为 'swarm_boxplot.png'")
plt.show()
```

## 💡 六、本节核心总结

| 步骤 | 工具 | 核心 API | 作用 |
|------|------|----------|------|
| 蜂群图 | Seaborn | `sns.swarmplot(x, y, hue)` | 不重叠散点，显示每个样本 |
| 箱线图叠加 | Seaborn | `sns.boxplot(..., boxprops)` | 中位数 + 四分位数叠加 |
| 透明度控制 | Matplotlib | `alpha=0.7` | 蜂群点半透明，不遮挡箱子 |
| 透明填充 | Matplotlib | `boxprops={"facecolor": "none"}` | 箱子不遮住底层散点 |
| 图层顺序 | Seaborn | 先 swarmplot 再 boxplot | 确保统计摘要在顶层 |
| 大数据替代 | Seaborn | `sns.stripplot(jitter=0.3)` | 数据量 > 500 时更快 |
